```python
# Load spam and non-spam training data
spam_file = open('spam.txt', 'r')
non_spam_file = open('non_spam.txt', 'r')
```

```python
# Tokenize words in spam and non-spam training data
spam_words = re.findall(r'\b\w+\b', spam_file.read().lower())
non_spam_words = re.findall(r'\b\w+\b', non_spam_file.read().lower())
```

```python
# Compute vocabulary of unique words in spam and non-spam training data
vocabulary = set(spam_words + non_spam_words)
```

```python
# Compute conditional probabilities of each word given spam or non-spam
spam_word_probs = {word: (spam_words.count(word) + 1) / (len(spam_words) + len(vocabulary)) for word in vocabulary}
non_spam_word_probs = {word: (non_spam_words.count(word) + 1) / (len(non_spam_words) + len(vocabulary)) for word in vocabulary}

# Define prior probability of spam and non-spam
prior_spam = 0.5
prior_non_spam = 1 - prior_spam
```

```python
def classify_email(email):
    # Tokenize words in email
    words = re.findall(r'\b\w+\b', email.lower())
    
    # Compute posterior probability of spam and non-spam using Bayes' rule
    spam_prob = prior_spam * math.prod(spam_word_probs.get(word, 1/(len(spam_words) + len(vocabulary))) for word in words)
    non_spam_prob = prior_non_spam * math.prod(non_spam_word_probs.get(word, 1/(len(non_spam_words) + len(vocabulary))) for word in words)
    evidence = spam_prob + non_spam_prob
    spam_posterior = spam_prob / evidence
    
    # Return posterior probability of spam
    return spam_posterior
```

```python
email = "Grab free food today! Free giveaway!"
spam_probability = classify_email(email)
print("Probability of being spam:", spam_probability)
```
